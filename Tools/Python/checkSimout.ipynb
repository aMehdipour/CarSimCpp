```python
plt.figure(figsize=(10, 6))
plt.plot(dfDense.Time, dfDense.State5)
plt.xlabel('Time (s)')
plt.ylabel('State 5')
plt.title('State 5 Over Time')
plt.grid()
plt.savefig('state5_over_time.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], df['deltaDeg'])
plt.xlabel('Time (s)')
plt.ylabel('Delta Degrees')
plt.title('Delta Degrees Over Time')
plt.grid(True)
plt.savefig('delta_degrees_over_time.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], df['slipAngleFrontDeg'])
plt.plot(df['Time'], df['slipAngleRearDeg'])
plt.legend(['Slip Angle Front', 'Slip Angle Rear'])
plt.xlabel('Time (s)')
plt.ylabel('Slip Angle (Degrees)')
plt.title('Slip Angle Over Time')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], df['Fyf'], label='Fyf')
plt.plot(df['Time'], df['Fyr'], label='Fyr')
plt.xlabel('Time (s)')
plt.ylabel('Lateral Force (N)')
plt.title('Lateral Forces Over Time')
plt.legend()
plt.grid(True)
plt.savefig('lateral_forces.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], df['beta'], label='Sideslip Angle')
plt.xlabel('Time (s)')
plt.ylabel('Sideslip Angle (Degrees)')
plt.title('Sideslip Angle Over Time')
plt.grid(True)
plt.savefig('sideslip_angle.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Time'], df['beta'], label='Sideslip Angle')
plt.xlabel('Time (s)')
plt.ylabel('Sideslip Angle (Degrees)')
plt.title('Sideslip Angle Over Time')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], df['betaDot'], label='Sideslip Rate')
plt.xlabel('Time (s)')
plt.ylabel('Sideslip Rate (Degrees/s)')
plt.title('Sideslip Rate Over Time')
plt.grid(True)
plt.savefig('sideslip_rate.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(df['Time'], df['rDot'], label='Yaw Rate')
plt.ylabel('Yaw Rate (Degrees/s)')
plt.title('Yaw Rate Over Time')
plt.grid(True)
plt.subplot(2, 1, 2)
plt.plot(df['Time'], df['Fyf'], label='Fyf')
plt.plot(df['Time'], df['Fyr'], label='Fyr')
plt.xlabel('Time (s)')
plt.ylabel('Lateral Force (N)')
plt.title('Lateral Forces Over Time')
plt.legend()
plt.grid(True)
plt.savefig('yaw_rate_vs_fy.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], df['vxBody'])
plt.xlabel('Time (s)')
plt.ylabel('Longitudinal Velocity (m/s)')
plt.title('Longitudinal Velocity Over Time')
plt.grid(True)
plt.savefig('longitudinal_velocity.png', dpi=300)
plt.show()
```
